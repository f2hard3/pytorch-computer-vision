```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
epoch = np.arange(1, 100)
train_loss = 1.5 / epoch
val_loss = .3+1 / epoch
```

```python
plt.figure(figsize=(10, 5))
plt.plot(train_loss, 'r-*')
plt.plot(val_loss, 'b-o ')
plt.legend(['train', 'validation'])
plt.title('MNIST')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```

```python
plt.figure(figsize=(5, 10))
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'ro', t, t**2, 'bs', t, 5*np.sin(t), 'g^')
plt.legend(['t', 't**2', '5tsin(t)'])
plt.show()
```

```python
names = ['First', 'Business', 'Economy']
prices = [20, 5, 1]

plt.figure(figsize=(15, 5))

plt.subplot(131)
plt.bar(names, prices)
plt.title(f'{names[0]}', fontsize=14)

plt.subplot(132)
plt.scatter(names, prices)
plt.xlabel(f'class', fontsize=12)

plt.subplot(133)
plt.plot(names, prices, linewidth=5.0)
plt.suptitle('Ticket Class', fontsize=14)
plt.axis('off')

plt.show()
```
