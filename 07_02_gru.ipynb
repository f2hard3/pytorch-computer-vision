```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/kospi.csv')

df.head()
```

```python
scaler = MinMaxScaler()
df[['Open', 'High', 'Low', 'Close', 'Volume']] = scaler.fit_transform(df[['Open', 'High', 'Low', 'Close', 'Volume']])

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
device = torch.device('cuda' if torch.cuda.is_available else 'cpu')

device
```

```python
X = df[['Open', 'High', 'Low', 'Volume']].values
y = df[['Close']].values
```

```python
def seq_data(x, y, sequence_length):    
    x_seq = []
    y_seq = []
    for i in range(len(x) - sequence_length):
        x_seq.append(x[i:i+sequence_length])
        y_seq.append(y[i+sequence_length])

    return torch.FloatTensor(np.array(x_seq)).to(device), torch.FloatTensor(np.array(y_seq)).to(device)
```

```python
split = 200
sequence_length = 5

x_seq, y_seq = seq_data(X, y, sequence_length)
x_train_seq = x_seq[:split]
y_train_seq = y_seq[:split]
x_test_seq = x_seq[split:]
y_test_seq = y_seq[split:]

x_train_seq.shape, y_train_seq.shape ,x_test_seq.shape, y_test_seq.shape
```

```python
trainset = TensorDataset(x_train_seq, y_train_seq)
testset = TensorDataset(x_test_seq, y_test_seq)

batch_size = 20
train_loader = DataLoader(dataset=trainset, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(dataset=testset, batch_size=batch_size, shuffle=False)
```

```python
input_size = x_seq.size(2)
num_layers = 2
hidden_size = 8
```

```python
class GRUModel(nn.Module):
    def __init__(self, input_size, hidden_size, sequence_length, num_layers, device) -> None:
        super().__init__()
        self.device = device
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm = nn.GRU(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size*sequence_length, 1)

    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size()[0], self.hidden_size).to(device)
        out, _ = self.lstm(x, h0)
        out = out.reshape(out.shape[0], -1)
        out = self.fc(out)
        
        return out
```

```python
model = GRUModel(
    input_size=input_size,
    hidden_size=hidden_size,
    sequence_length=sequence_length,
    num_layers=num_layers,
    device=device
)
model = model.to(device)
```

```python
criterion = nn.MSELoss()

lr = 1e-03
num_epochs = 200
optimizer = optim.Adam(params=model.parameters(), lr=lr)
```

```python
model.train()
loss_graph = []
n = len(train_loader)
for epoch in range(num_epochs):
    running_loss = 0.0

    for data in train_loader:
        seq, target = data
        out = model(seq)
        loss = criterion(out, target)
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        running_loss += loss.item()

    loss_graph.append(running_loss/n)

    if epoch % 100 == 0:
        print(f'[epoch: {epoch}] loss: {running_loss/n:.4f}')
```

```python
plt.figure(figsize=(20, 10))
plt.plot(loss_graph)
plt.show()
```

```python
def plotting(train_loader, test_loader, actual):
    with torch.inference_mode():
        train_pred = []
        test_pred = []
        for data in train_loader:
            seq, _ = data
            out = model(seq)
            train_pred += out.cpu().numpy().tolist()

        for data in test_loader:
            seq, _ = data
            out = model(seq)
            test_pred += out.cpu().numpy().tolist()
    
    total = train_pred+test_pred

    plt.figure(figsize=(20, 10))
    plt.plot(np.ones(100)*len(train_pred), np.linspace(0, 1, 100))
    plt.plot(actual, '--')
    plt.plot(total, 'b', linewidth=0.6)

    plt.legend(['train boundary', 'actual', 'prediction'])

    plt.show()
        
```
